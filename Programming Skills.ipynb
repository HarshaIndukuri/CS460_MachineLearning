```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#a
m = [0,0]
cov = [(1,0),(0,1)] #identity covariance matrix
d = np.random.multivariate_normal(m,cov,100)

plt.scatter(d[:,0],d[:,1],c = 'blue')
```

```python
#b
m2 = [-1,1]
d2 = np.random.multivariate_normal(m2,cov,100)

plt.scatter(d2[:,0],d2[:,1],c = 'blue')
```

Points move together from around the origin to around (-1,1)

```python
#c
cov2 = [(2,0),(0,2)]#doubling variance of each variable
d3 = np.random.multivariate_normal(m,cov2,100)

plt.scatter(d3[:,0],d3[:,1],c = 'blue')
```

Points seem to be slightly more spread out, but not much different from identity covariance matrix

```python
#d
cov3 = [(1,0.5),(0.5,1)]#doubling variance of each variable
d4 = np.random.multivariate_normal(m,cov3,100)

plt.scatter(d4[:,0],d4[:,1],c = 'blue')
```

Points are clearly scattered from bottom left to top right and not in all directions

```python
#e
cov4 = [(1,-0.5),(-0.5,1)]#doubling variance of each variable
d5 = np.random.multivariate_normal(m,cov4,100)

plt.scatter(d5[:,0],d5[:,1],c = 'blue')
```

Points are clearly scatterd from top right to bottom left
