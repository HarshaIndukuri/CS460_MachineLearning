```python
import math
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
```

# KMeans and DBSCAN Using Scikit Learn on simply distinguishable clusters

```python
#Creating 4 clusters in 2d space with normal and uniform
X = np.random.normal(10,0.5,200)
X1 = np.random.uniform(2,3,50)
X2 = np.random.normal(4,0.5,50)
X3 = np.random.uniform(7,8,50)
X[50:100] = X1
X[100:150] = X2
X[150:200] = X3

Y = np.random.normal(1,1,200)
Y1 = np.random.uniform(2,3,50)
Y2 = np.random.normal(9,0.5,50)
Y3 = np.random.uniform(3,4,50)
Y[50:100] = Y1
Y[100:150] = Y2
Y[150:200] = Y3

D = np.vstack((X,Y)).T #combining X and Y
```

```python
#Plotting Our Random Data
plt.scatter(X,Y,c = 'black')
plt.xlabel('X')
plt.ylabel('Y')
```

# 1.KMeans

```python
#Using Sklearn Kmeans to cluster our data
model = KMeans(n_clusters = 4)
model.fit(D)
centroids = model.cluster_centers_
labels = model.labels_
```

```python
#Now we plot our data with the centroids scikit learn's kmeans converged upon
sns.scatterplot(x=X,y=Y,hue=labels,palette=['green','red','purple','dodgerblue'])
sns.scatterplot(x=centroids[:,0],y=centroids[:,1],sizes = [10,10],color = 'black')
```

# 2.DBSCAN

```python
#DBSCAN on fairly simple clusters
model4 = DBSCAN(eps=1.5,min_samples=4)
model4.fit(D)
labels4 = model4.labels_
```

```python
sns.scatterplot(x=X,y=Y,hue=labels4,palette=['green','red','purple','blue'])
```

# KMeans and DBSCAN using Scikit Learn on Semicircular Data

```python
#Now we create a new dataset to compare dbscan and scikit learn, we will try semicircular data

C = np.zeros(1000)
S = np.zeros(1000)
p = math.pi

for i in range(1000):
    if i>=0 and i<=249:
        r = np.random.normal(1,0.1)
        theta = np.random.uniform(0,1.05)
        C[i] = r*math.cos(theta*p)
        S[i] = r*math.sin(theta*p)
    
    elif i>=250 and i<=499:
        r = np.random.normal(3.0,0.1)
        theta = np.random.uniform(0,1.05)
        C[i] = r*math.cos(theta*p)
        S[i] = r*math.sin(theta*p)

    elif i>=500 and i<=749:
        r = np.random.normal(5,0.1)
        theta = np.random.uniform(0,1.05)
        C[i] = r*math.cos(theta*p)
        S[i] = r*math.sin(theta*p) 
    
    else :
        r = np.random.normal(7,0.1)
        theta = np.random.uniform(0,1.05)
        C[i] = r*math.cos(theta*p)
        S[i] = r*math.sin(theta*p)

D2 = np.vstack((C,S)).T
```

```python
#plotting our initial data
sns.scatterplot(x=C,y=S,color = 'black')
```

# 1.KMeans 

```python
#Kmeans on circular data
model2 = KMeans(n_clusters = 4)
model2.fit(D2)
centroids2 = model2.cluster_centers_
labels2 = model2.labels_
```

```python
#Now we plot our data with the centroids scikit learn's kmeans converged upon
sns.scatterplot(x=C,y=S,hue=labels2,palette=['green','red','purple','dodgerblue'])
sns.scatterplot(x=centroids2[:,0],y=centroids2[:,1],sizes = [10,10],color = 'black')
```

# 2.DBSCAN

```python
#DBSCAN on circular data
model3 = DBSCAN(eps=1,min_samples=4)
model3.fit(D2)
labels3 = model3.labels_
```

```python
#plotting our data based on cluster created by DBSCAN
sns.scatterplot(x=C,y=S,hue=labels3,palette=['green','red','purple','dodgerblue'])
```
